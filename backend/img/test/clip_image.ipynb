```python
# 필요한 라이브러리 임포트
import os
from PIL import Image
from stability_sdk import client
import stability_sdk.interfaces.gooseai.generation.generation_pb2 as generation
from transformers import CLIPProcessor, CLIPModel
import io

# Stability AI API 초기화
os.environ['STABILITY_KEY'] = 'sk-2N8ZSE6aMkRCxTij0mmPOz6BKIdHvH1NzFCnzeetQkNISfjy'  # Stability AI API 키 입력
stability_api = client.StabilityInference(
    key=os.environ['STABILITY_KEY'],
    verbose=True,
)

# CLIP 모델 및 프로세서 로드
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Stable Diffusion API 이미지 생성 함수 정의
def generate_image(prompt, seed=992446758, width=512, height=512, steps=50, cfg_scale=10.0):
    answers = stability_api.generate(
        prompt=prompt,
        seed=seed,
        steps=steps,
        cfg_scale=cfg_scale,
        width=width,
        height=height,
        samples=1,
        sampler=generation.SAMPLER_K_DPMPP_2M,
    )
    for resp in answers:
        for artifact in resp.artifacts:
            if artifact.type == generation.ARTIFACT_IMAGE:
                img = Image.open(io.BytesIO(artifact.binary))
                return img

# CLIP 모델로 이미지 특징 추출 함수 정의
def extract_image_features(image):
    inputs = clip_processor(images=image, return_tensors="pt")
    outputs = clip_model.get_image_features(**inputs)
    return outputs

# Scene prompts 정의
scene_prompts = [
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. A scenic rural Korean village under soft, warm sunlight, a young boy named Minsu runs towards his grandfather’s traditional Korean house (hanok), surrounded by green grass and blooming flowers. High-quality, wide-angle shot, soft lighting, ambient sunlight.",
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. Minsu warmly hugs his grandfather with a soft, happy expression. The grandfather, with gray hair and a gentle smile, sits with him in a cozy, warm-toned living room decorated with traditional Korean wooden furniture. Soft sunlight filters through a nearby window, creating a serene and homely atmosphere. Medium shot, ambient lighting, high detail.",
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. Grandfather recalls his childhood during the Korean War in a nostalgic black-and-white scene. A faded photo appears, showing a collapsed bridge with refugees gathered around, capturing the somber reality. High-detail monochrome image with a focus on emotional expression.",
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. Grandfather, smiling softly, tells Minsu about a mischievous story of eating unripe grapes and suffering a stomachache. Minsu listens with a worried look on his face. Shot in warm lighting with a cinematic feel, close-up capture on the grandfather’s and Minsu’s expressions.",
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. A iknd elderly man in Chungcheong Province shares barley and potatoes with a young grandfather. The elderly man's warm smile and the gratitude in the young grandfather's eyes are captured in the scene. Soft, ambient lighting, medium shot with a nostalgic, heartfelt tone.",
    "Illustrated in a refined and cohesive style inspired by modern Korean children's storybooks, featuring soft gradients, clean lines, vibrant yet harmonious colors, and a warm, whimsical atmosphere. As the story ends, Minsu sits by his grandfather’s side, touched by the stories he’s heard. Both sit together under soft, warm sunlight, Minsu nodding thoughtfully. Wide-angle shot with natural lighting, capturing the warm bond between them.",
]

# 기본 설정
seed=992446758  # 고정된 seed 값
updated_prompts = []  # 업데이트된 프롬프트 저장

# 각 Scene의 이미지 생성 및 CLIP 특징 추출
for i, prompt in enumerate(scene_prompts):
    print(f"Generating Image for Scene {i + 1}...")

    # 이미지 생성
    generated_image = generate_image(prompt=prompt, seed=seed + i)
    generated_image.save(f"scene_{i + 1}_image.png")
    print(f"Image for Scene {i + 1} saved as 'scene_{i + 1}_image.png'")

    # Colab에서 이미지 표시
    print(f"\nScene {i + 1} Image:")
    display(generated_image)

    # 이미지 특징 추출
    image_features = extract_image_features(generated_image)
    print(f"Extracted features for Scene {i + 1}: {image_features}")

    # 프롬프트 업데이트
    updated_prompt = prompt + f", reflecting the style of the previous image with features: {image_features}"
    updated_prompts.append(updated_prompt)

# 모든 업데이트된 프롬프트 출력
print("\nUpdated Prompts:")
for i, updated_prompt in enumerate(updated_prompts):
    print(f"Scene {i + 1}: {updated_prompt}")
```
